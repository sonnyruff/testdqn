```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.cm as cm
import numpy as np

sns.set(style="whitegrid")

csv_path = "wandb_data/2 narrow sweep/"
```

```python
df1 = pd.read_csv(csv_path + "1 Con NN init_std 0 2.csv")
df2 = pd.read_csv(csv_path + "2 Con NN init_std 3 5.csv")
df3 = pd.read_csv(csv_path + "3 Con NN init_std 6 9.csv")
_df1 = pd.concat([df1, df2, df3], ignore_index=True)
```

```python
_df1.info()
```

```python
x_n = 'noisy_layer_init_std'
y_n = 'regret'

# plt.figure(figsize=(10, 7))
plt.figure(figsize=(10, 5))

sns.lineplot(
    data=_df1,
    x=x_n,
    y=y_n,
    hue='env_id',
    errorbar='se',
    style='env_id',
    markers=True,
    palette='colorblind',
)

plt.xlabel(x_n)
plt.ylabel(y_n)
plt.title(f'{y_n} vs. {x_n}')
plt.grid(True)
# plt.legend(title='Environment', loc='center right')
plt.legend(title='Environment', loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout()

plt.savefig("results/regret_vs_noisy_std 1.png", dpi=300, bbox_inches='tight')

plt.show()
```

```python
df4 = pd.read_csv(csv_path + "4 MNIST init_std 0.csv")
df5 = pd.read_csv(csv_path + "5 MNIST init_std 1.csv")
df6 = pd.read_csv(csv_path + "6 MNIST init_std 2.csv")
_df2 = pd.concat([df4, df5, df6], ignore_index=True)
```

```python
x_n = 'noisy_layer_init_std'
y_n = 'regret'

# plt.figure(figsize=(10, 7))
plt.figure(figsize=(10, 5))

sns.lineplot(
    data=_df2,
    x=x_n,
    y=y_n,
    hue='env_id',
    errorbar='se',
    marker='o',
    palette='colorblind'
)

plt.xlabel(x_n)
plt.ylabel(y_n)
plt.title(f'{y_n} vs. {x_n}')
plt.grid(True)
# plt.legend(title='Environment', loc='center right')
plt.legend(title='Environment', loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout()

plt.savefig("results/regret_vs_noisy_std 2.png", dpi=300, bbox_inches='tight')

plt.show()
```

```python
df7 = pd.read_csv(csv_path + "7 MNIST init_std 2.csv")
df8 = pd.read_csv(csv_path + "8 MNIST init_std 0.csv")
df9 = pd.read_csv(csv_path + "9 MNIST init_std 1.csv")
_df3 = pd.concat([_df2, df7, df8, df9], ignore_index=True)
# _df3
```
