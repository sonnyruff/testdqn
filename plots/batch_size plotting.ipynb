```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.cm as cm
import numpy as np

sns.set(style="whitegrid")

csv_path = "wandb_data/batch_size_sweep/"
```

```python
df = pd.read_csv(csv_path + "big_sweep_200.csv")
# df.info()
```

```python
df = df[df.memory_size == 1000]
# df.info()
```

```python
x_name = 'batch_size'
y_name = 'mean_regret'

# envs = df['env_id'].unique()
# colors = plt.cm.tab10(np.linspace(0, 1, len(envs)))
# color_map = dict(zip(envs, colors))

# plt.figure(figsize=(10, 5))

# for (env, seed), group_df in df.groupby(['env_id', 'seed']):
#     plt.plot(group_df[x_name],group_df[y_name],label=f'{env} - Seed {seed}',color=color_map[env])

# plt.title('Mean Regret vs Batch Size')
# plt.xlabel(x_name)
# plt.ylabel(y_name)
# plt.grid(True)
# plt.legend(title='Env - Seed', loc='upper left')
# plt.tight_layout()
# plt.show()
```

```python
df500 = pd.read_csv(csv_path + "batch_size_sweep_500.csv")
df1000 = pd.read_csv(csv_path + "batch_size_sweep_1000.csv")
df1500 = pd.read_csv(csv_path + "batch_size_sweep_1500.csv")
df2000 = pd.read_csv(csv_path + "batch_size_sweep_2000.csv")
df = pd.concat([df500, df1000, df1500, df2000], ignore_index=True)
plt.figure(figsize=(10, 7))

sns.lineplot(
    data=df[df.memory_size.isin([500, 1000, 1500, 2000])],
    x=x_name,
    y='mean_regret',
    hue='env_id',
    style='memory_size',
    errorbar='se',
    palette='colorblind'
)

plt.title('Mean Regret vs Batch Size (±1 std) - batch size 200')
plt.xlabel(x_name)
plt.ylabel('mean_regret')
plt.grid(True)
plt.legend(title='Environment', loc='center right')
plt.tight_layout()

plt.savefig("results/mean_regret_vs_batch__stationary.png", dpi=300, bbox_inches='tight')

plt.show()
```

```python
df = pd.read_csv(csv_path + "big_sweep_200.csv")
plt.figure(figsize=(10, 7))

sns.lineplot(
    data=df[df.memory_size.isin([500, 1000, 1500, 2000])],
    x=x_name,
    y='mean_regret',
    hue='env_id',
    style='memory_size',
    errorbar='se',
    palette='colorblind'
)

plt.title('Mean Regret vs Batch Size (±1 std) - Dynamic Rate 200')
plt.xlabel(x_name)
plt.ylabel('mean_regret')
plt.grid(True)
plt.legend(title='Environment', loc='center right')
plt.tight_layout()

plt.savefig("results/mean_regret_vs_batch_size__dynrate200.png", dpi=300, bbox_inches='tight')

plt.show()
```
